```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Datos experimentales
Datos = {
    'T,s': [0,443,545,767,895,1013,1100,1211,1312,1412,1545,1630,1704,1793,1880,1944,2013,2142,2225,
            2385,2537,2645,2907,3247,3428,3634,4023,4283,4555,4923,5327,5977],
    'Temp condensador, °C': [23.2,55.9,59.0,60.9,62.4,63.3,64.0,64.2,64.4,64.6,65.0,65.2,65.3,65.3,65.4,
                             65.5,65.6,65.6,65.5,65.4,65.2,65.0,64.9,64.8,65.0,65.1,64.8,65.2,65.1,65.0,63.3,60.7],
    'Temp caldera, °C': [57.5,65.4,67.7,69.5,70.7,71.8,72.9,73.1,73.8,74.2,74.5,74.7,74.8,75.0,75.1,75.2,75.2,
                         75.4,75.7,76.1,76.2,76.6,77.2,78.3,78.8,80.0,81.1,82.4,83.9,85.1,90.4,96.9],
    'V, L': [0.00,0.00,0.50,1.00,1.50,2.00,2.50,3.00,3.50,4.00,4.50,5.00,5.50,6.00,6.50,7.00,7.50,8.00,
             8.50,9.00,9.50,10.00,10.50,11.00,11.50,12.00,12.50,13.00,13.50,14.00,14.50,14.60],
}

df_datos = pd.DataFrame(Datos)
display

# Datos de densidades

densidades_data = {
    'kg/m³': {'Agua': 983.2, 'Hexano': 660.6},
    'kg/L': {'Agua': 0.9832, 'Hexano': 0.6606},
}
df_densidades = pd.DataFrame(densidades_data)

Vhexano_caldera = 14.6 - df_datos['V, L']
mhexano_caldera = Vhexano_caldera * df_densidades.loc['Hexano', 'kg/L']
Vagua_caldera = 12.0
magua_caldera = Vagua_caldera* df_densidades.loc['Agua', 'kg/L']
fraccion_masica_hexano = mhexano_caldera / (mhexano_caldera + magua_caldera)

df_calculos = pd.DataFrame({
    'T,s': df_datos['T,s'],
    'Vhexano_caldera, L': Vhexano_caldera,
    'mhexano_caldera, kg': mhexano_caldera,
    'magua_caldera, kg': magua_caldera,
    'fraccion_masica_hexano': fraccion_masica_hexano,
})
display(df_calculos)
# Mostrar DataFrame de densidades

display(df_densidades)

datos_semillas = {
    'Peso inicial, g': 2000,
    'Riqueza grasa %': 30.48, 
    'aceite recuperado, g': 450
}
df_semillas = pd.DataFrame(datos_semillas, index=['Semillas'])
display(df_semillas)

plt.plot(fraccion_masica_hexano, df_datos['Temp caldera, °C'], 'o')
plt.xlabel('Fracción masica de hexano')
plt.ylabel('Temperatura de la caldera, °C')
plt.title('Temperatura de la caldera vs fracción masica de hexano')
plt.grid()
plt.show()

aceite_pipas = df_semillas['Riqueza grasa %'] * df_semillas['Peso inicial, g'] / 100 # para tener el peso en kg y el porcentaje en fracción
Rendimiento = df_semillas['aceite recuperado, g'] / aceite_pipas
df_semillas['aceite pipas, kg'] = aceite_pipas
df_semillas['Rendimiento'] = Rendimiento.map('{:.2%}'.format)
display(df_semillas)


Vhexano_extraido = np.array(Vhexano_caldera[::-1])

Qhexano_values = []
# Calculamos Qhexano para cada intervalo de tiempo
for i in range(len(Vhexano_extraido)):
    if i == 0:
        Qhexano = 0  # Asignar Qhexano = 0 para el primer elemento
    else:
        delta_V = Vhexano_extraido[i] - Vhexano_extraido[i-1]
        delta_T = df_datos['T,s'][i] - df_datos['T,s'][i-1]
        
        # Evitar división por cero
        if delta_T != 0:
            Qhexano = delta_V / delta_T
        else:
            Qhexano = 0  
    Qhexano_values.append(Qhexano)
mmovil =pd.Series(Qhexano_values).rolling(window=2).mean()
mmovil2 =pd.Series(Qhexano_values).rolling(window=3).mean()

df_Qhexano = pd.DataFrame({
    'T,s': df_datos['T,s'],
    'Qhexano, L/s': Qhexano_values,
})
display(df_Qhexano)

plt.scatter(df_datos['T,s'], Qhexano_values, label='Qhexano', color='blue')
plt.plot(df_datos['T,s'], mmovil, '--', color='red', label='Media movil 2 valores', linewidth=2)
plt.plot(df_datos['T,s'], mmovil2, color='green', label='Media movil 3 valores', linewidth=2, alpha=0.5)
plt.ylabel('Qhexano, L/s')
plt.xlabel('Tiempo, s')
plt.title('Qhexano vs tiempo')
plt.grid()
plt.legend()
plt.show()
```
