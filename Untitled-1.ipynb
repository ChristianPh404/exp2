```python
import numpy as np
import matplotlib.pyplot as plt
from CoolProp.CoolProp import PropsSI

# Definir el refrigerante y condiciones de operacin
refrigerante = 'R134a'
T_evaporador = 278.15  # Temperatura en evaporador (K), por ejemplo, 5C
T_condensador = 313.15  # Temperatura en condensador (K), por ejemplo, 40C
sobrecalentamiento = 5  # Temperatura de recalentamiento en Kelvin (ej. 5 K por encima de T_evaporador)

# Puntos del ciclo
# 1. Punto h1: Evaporacin (vapor saturado a baja presin)
P_evaporacion = PropsSI('P', 'T', T_evaporador, 'Q', 1, refrigerante)
h1 = PropsSI('H', 'T', T_evaporador, 'Q', 1, refrigerante) / 1000  # en kJ/kg
s1 = PropsSI('S', 'T', T_evaporador, 'Q', 1, refrigerante)  # Entropa en J/kgK

# 2. Punto h': Recalentamiento (vapor sobrecalentado)
T_recalentado = T_evaporador + sobrecalentamiento
h_recalentado = PropsSI('H', 'T', T_recalentado, 'P', P_evaporacion, refrigerante) / 1000  # en kJ/kg

# 3. Punto h2: Compresin (vapor sobrecalentado a alta presin)
P_condensacion = PropsSI('P', 'T', T_condensador, 'Q', 1, refrigerante)
s2 = s1  # Proceso isentrpico (misma entropa)
h2 = PropsSI('H', 'P', P_condensacion, 'S', s2, refrigerante) / 1000  # en kJ/kg

# 4. Punto h3: Condensacin (lquido saturado a alta presin)
h3 = PropsSI('H', 'T', T_condensador, 'Q', 0, refrigerante) / 1000  # en kJ/kg

# 5. Punto h4: Expansin (mezcla de lquido y vapor a baja presin)
h4 = h3  # En la expansin se asume entalpa constante (proceso isoentlpico)

# Generar datos de la curva de saturacin para el diagrama p-h
presiones = np.linspace(100000, 6000000, 200)  # 0.1 MPa a 2 MPa
h_liquido = []
h_vapor = []
p_saturacion = []


for T in np.linspace(230, 350, 200):  # Temperatura en Kelvin
    try:
        h_fase_liquido = PropsSI('H', 'T', T, 'Q', 0, refrigerante)
        h_fase_vapor = PropsSI('H', 'T', T, 'Q', 1, refrigerante)
        p_sat = PropsSI('P', 'T', T, 'Q', 0, refrigerante)
        
        h_liquido.append(h_fase_liquido / 1000)  # Convertir a kJ/kg
        h_vapor.append(h_fase_vapor / 1000)      # Convertir a kJ/kg
        p_saturacion.append(p_sat / 1000)        # Convertir a kPa
    except:
        pass  # Ignorar errores fuera del rango del refrigerante

# Graficar el ciclo en el diagrama p-h
plt.figure(figsize=(10, 6))
plt.plot(h_liquido, p_saturacion, label="Lquido saturado", color='blue')
plt.plot(h_vapor, p_saturacion, label="Vapor saturado", color='red')
plt.yscale('log')

# Marcar los puntos del ciclo
plt.plot([h1, h_recalentado, h2, h3, h4, h1], 
         [P_evaporacion/1000, P_evaporacion/1000, P_condensacion/1000, P_condensacion/1000, P_evaporacion/1000, P_evaporacion/1000], 
         marker='o', color='green', label="Ciclo de refrigeracin")
plt.text(h1, P_evaporacion/1000, 'h1: Evaporador', ha='right')
plt.text(h_recalentado, P_evaporacion/1000, "h': Recalentamiento", ha='right')
plt.text(h2, P_condensacion/1000, 'h2: Compresor', ha='right')
plt.text(h3, P_condensacion/1000, 'h3: Condensador', ha='right')
plt.text(h4, P_evaporacion/1000, 'h4: Expansin', ha='right')

# Dibujar las lneas de recalentamiento
plt.plot([h1, h_recalentado], [P_evaporacion/1000, P_evaporacion/1000], color='purple', linestyle="--", label="Lnea de recalentamiento")
plt.plot([h_recalentado, h2], [P_evaporacion/1000, P_condensacion/1000], color='purple', linestyle="--")

# Configuracin del grfico
plt.xlabel("Entalpa especfica (kJ/kg)")
plt.ylabel("Presin (kPa)")
plt.title(f"Diagrama p-h para {refrigerante} con Ciclo de Refrigeracin")
plt.legend()
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()
```
